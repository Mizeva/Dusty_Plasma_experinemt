```python
from numpy.core.function_base import linspace
import matplotlib.pyplot as plt
import numpy as np
from statistics import mean
import math
from scipy.optimize import curve_fit
import pandas as pd
from scipy.misc import derivative
from statistics import mean
```

```python
f = np.loadtxt("/content/Y1000.txt")
par = np.array([[0.0 for i in range(len(f))] for j in range(4)])
for i in range(len(f)):
    for j in range(0, 4):
        par[j][i] = f[i][j]

f1 = np.loadtxt("/content/X1000.txt")
x = np.array([[0.0 for i in range(len(f1))] for j in range(4)])
for i in range(len(f1)):
    for j in range(0, 4):
        x[j][i] = f1[i][j]

l = 4000
r = l + 40

c = ['r', 'g', 'royalblue', 'orange']

for i in range(4):
  print(mean(par[i]))
  plt.plot(x[i][l:r], par[i][l:r], color = c[i])

#plt.plot(x[0][l:r], par[0][l:r], color = 'black')
plt.grid()
plt.xlabel("x, пикс")
plt.ylabel("y, пикс")
plt.show()
len(f)
#до 8400 нормально
```
